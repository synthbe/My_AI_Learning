```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.15,
    height_shift_range=0.15,
    shear_range=0.20,
    zoom_range=0.1,
    channel_shift_range=10.,
    horizontal_flip=True
)
```

```python
directory = '../../datasets/brain_tumors/normal/'
chosen_image = random.choice(os.listdir(directory))
image_path = directory + chosen_image
```

```python
image_path
```

```python
assert os.path.isfile(image_path)
```

```python
original_image = plt.imread(image_path)
plt.imshow(original_image)
```

```python
original_image = original_image.reshape(1, 256, 256, 3)
```

```python
aug_iter = datagen.flow(original_image)
```

```python
aug_images = [next(aug_iter)[0].astype(np.uint8) for _ in range(10)]
```

```python
fig, aux = plt.subplots(nrows=2, ncols=5, figsize=(10, 5))
i, j = 0, 0
for image in aug_images:
    ax = aux[i, j]
    ax.imshow(image)
    j = j + 1 if i >= 1 else j
    i = i + 1 if i < 1 else 0
plt.show()
```
